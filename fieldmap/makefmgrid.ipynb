```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
def tendig(n):
    digit_count = len(str(n))
    if digit_count > 9:
        n = round(n,7)
        return n
    else:
        return n
```

```python
# 0 1 2 3  4  5
# x y z Bx By Bz

# upload the xyz-data file
file = 'fmgrid.txt'
data = np.loadtxt(file)

# Arrange values from detectors into arrays:
x_vals = []; y_vals = []; z_vals = []
Bx_vals = []; By_vals = []; Bz_vals =[]; B_vals = []
id_vals = []

for i in range(data.shape[0]):
    row = data[i]
    id_vals.append(i)

    x = row[0]; y = row[1]; z = row[2]
    x_vals.append(x/10)
    y_vals.append(y/10)
    z_vals.append(z/10)
    Bx = row[3]; By = row[4]; Bz = row[5]
    Bx_vals.append(Bx)
    By_vals.append(By)
    Bz_vals.append(Bz)

    Btot = (Bx**2 + By**2 + Bz**2)**(1/2)
    B_vals.append(Btot)
```

```python
# Combine the data into a list of tuples
data_combined = list(zip(z_vals, y_vals, x_vals, Bx_vals, By_vals, Bz_vals))

# Sort the data: primary key z, secondary y, tertiary x
data_sorted = sorted(data_combined)

# Unpack the sorted data
z_sorted, y_sorted, x_sorted, Bx_sorted, By_sorted, Bz_sorted = zip(*data_sorted)

# Optionally convert back to lists
x_ordered = list(x_sorted)
y_ordered = list(y_sorted)
z_ordered = list(z_sorted)
Bx_ordered = list(Bx_sorted)
By_ordered = list(By_sorted)
Bz_ordered = list(Bz_sorted)
```

```python
# This block is to test the output

# Write the content into an output file in the format:
# x[cm] y[cm] z[cm] Bx[T] By[T] Bz[T]

output = 'test.txt'


with open(output, "w+") as file:
    for i in range(1000):
        x = str(tendig(x_ordered[i]))
        y = str(tendig(y_ordered[i]))
        z = str(tendig(z_ordered[i]))
        Bx = str(tendig(Bx_ordered[i]))
        By = str(tendig(By_ordered[i]))
        Bz = str(tendig(Bz_ordered[i]))

        if i==0:
            line = f"MGNDATA   {x:>10}{y:>10}{z:>10}{Bx:>10}{By:>10}{Bz:>10}FMGRID\n"
        elif i==1:
            line = f"MGNDATA   {x:>10}{y:>10}{z:>10}{Bx:>10}{By:>10}{Bz:>10} &\n"
        else:
            line = f"MGNDATA   {x:>10}{y:>10}{z:>10}{Bx:>10}{By:>10}{Bz:>10} &&\n"

        file.write(line)
        #print(line)
```
