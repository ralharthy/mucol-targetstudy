```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
def tendig(n):
    digit_count = len(str(n))
    if digit_count > 9:
        n = round(n,7)
        return n
    else:
        return n
```

```python
# 0 1 2  3
# r z Br Bz

# upload the rz-data file
file = 'fmcylinder.txt'
data = np.loadtxt(file)
```

```python
# Arrange values from detectors into arrays:
r_vals = []; z_vals = []
Br_vals = []; Bz_vals =[]; B_vals = []

for i in range(data.shape[0]):
    row = data[i]

    r = row[0]; z = row[1]
    r_vals.append(r/10)
    z_vals.append(z/10)
    Br = row[2]; Bz = row[3]
    Br_vals.append(Br)
    Bz_vals.append(Bz)

    Btot = (Br**2 + Bz**2)**(1/2)
    B_vals.append(Btot)
```

```python
# Combine the data into a list of tuples
data_combined_1 = list(zip(z_vals, r_vals, Br_vals, Bz_vals))

# Sort the data: primary key z, secondary y, tertiary x
data_sorted_1 = sorted(data_combined_1)

# Unpack the sorted data
z_sorted, r_sorted, Br_sorted, Bz_sorted = zip(*data_sorted_1)

# Optionally convert back to lists
r_ordered = list(r_sorted)
z_ordered = list(z_sorted)
Br_ordered = list(Br_sorted)
Bz_ordered = list(Bz_sorted)
```

```python
# This block is to test the output

# Write the content into an output file in the format:
# r[cm] z[cm] Br[T] Bz[T]

output2 = 'testrz.txt'


with open(output2, "w+") as file:
    for i in range(1000):
        r = str(tendig(r_ordered[i]))
        z = str(tendig(z_ordered[i]))
        Br = str(tendig(Br_ordered[i]))
        Bz = str(tendig(Bz_ordered[i]))

        if i % 2 == 0:
            line = f"MGNDATA   {r:>10}{z:>10}{Br:>10}{Bz:>10} |"
        else:
            if i == 1:
                line = f"{r:>10}{z:>10}{Br:>10}{Bz:>10}FMGRID\n"
            elif i == 3:
                line = f"{r:>10}{z:>10}{Br:>10}{Bz:>10} &\n"
            else:
                line = f"{r:>10}{z:>10}{Br:>10}{Bz:>10} &&\n"

        file.write(line)
        #print(line)  
```
